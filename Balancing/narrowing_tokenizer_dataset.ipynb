```python
from typing import List
import pandas as pd 
import numpy as np
```

```python
DATASET_FILENAME: str = r"D:\Obfuscation\data\datasets\TOKENIZER_DATASET.csv"
UPDATED_DATASET_FILENAME: str = r"D:\Obfuscation\data\datasets\TOKENIZER_DATASET_UPDATED.csv"
DROP_FEATURES_FILENAME: str = r"D:\Obfuscation\data\drop_features.txt"
```

```python
def get_drop_features(filename: str) -> List[str]:
    with open(filename, 'r') as file:
        raw_features: List[str] = file.readlines()
    features: List[str] = list(map(lambda feature: str.strip(feature, '\n'), raw_features))
    return features
```

```python
drop_features = get_drop_features(DROP_FEATURES_FILENAME)
```

```python
len(drop_features)
```

```python
obfuscated_commands = pd.read_csv(DATASET_FILENAME)
```

```python
obfuscated_commands.head()
None
```

```python
deleted_features: List[str] = ["command_clear", "command_obfuscated"] + drop_features
```

```python
updated_df = obfuscated_commands.drop(columns=deleted_features)
```

```python
updated_df.to_csv(UPDATED_DATASET_FILENAME, index=False, sep=';')
```

```python
import csv
from typing import List, Dict
```

```python
def get_data(filename: str) -> list:
    with open(filename, encoding="UTF8") as file:
        reader = csv.reader(file)
        for row in reader:
            yield row
```

```python
def process_selection_methods(filename: str) -> Dict[str, List[str]]:  # feature: list of methods, which have chosen this feature
    def init_result(features: List[str]) -> Dict[str, List[str]]:
        result: Dict[str, List[str]] = {}
        for feature in features:
            result[feature] = []
        return result
    
    generator = get_data(filename)
    header: List[str] = next(generator)
    features: List[str] = header[1:]
    result: Dict[str, List[str]] = init_result(features)
    while True:
        row: str = next(generator, None)
        if row is None:
            break
        selection_method: str = row[0]
        feature_values: List[int] = [int(value) for value in row[1:]]
        for i in range(len(feature_values)):
            if feature_values[i] == 1:
                feature: str = features[i]
                result[feature].append(selection_method)
    return result
```

```python
filename: str = r"D:\Obfuscation\data\selection_result.csv"
result: Dict[str, List[str]] = process_selection_methods(filename)
sorted_result: Dict[str, List[str]] = dict(sorted(result.items(), key=lambda pair: len(pair[1]), reverse=True))
```

```python
for feature, methods in sorted_result.items():
    print(f"Признак '{feature}': количество методов = {len(methods)}.\n")
```

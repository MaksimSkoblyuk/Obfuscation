```python
drop_features_prefixes: List[str] = [
    "AstArrayElementMetrics_CharacterDistribution",
    "AstGroupedAssignmentStatements",
    "AstGroupedAstTypes",
    "AstGroupedBinaryExpressionOperators",
    "AstCmdletMetrics_CharacterDistribution",
    "AstCommandParameterNameMetrics_CharacterDistribution",
    "AstCommentMetrics_CharacterDistribution",
    "AstCommentMetrics",
    "AstConvertExpressionMetrics_CharacterDistribution",
    "AstFunctionNameMetrics_CharacterDistribution",
    "AstIntegerAndDoubleMetrics_CharacterDistribution",
    "AstIntegerAndDoubleMetrics",
    "AstInvocationOperatorInvokedObjectMetrics_CharacterDistribution",
    "LineByLineMetrics_CharacterDistribution",
    "LineByLineMetrics",
    "AstMemberArgumentMetrics_CharacterDistribution",
    "AstMemberMetrics_CharacterDistribution",
    "AstStringMetrics_CharacterDistribution",
    "AstTypeConstraintMetrics_CharacterDistribution",
    "AstTypeExpressionMetrics_CharacterDistribution",
    "AstTypeExpressionMetrics",
    "AstGroupedUnaryExpressionOperators",
    "AstVariableNameMetrics_CharacterDistribution",
]
```

```python
def write_file(filename: str, value: str) -> None:
    with open(filename, mode='a') as file:
        file.write(f"{value}\n")
```

```python
# calculate features with these prefixes
def get_features_quantity(features: List[str], prefixes: List[str]) -> Dict[str, int]:
    pref_ptr = feature_ptr = 0
    was_previously: bool = False  # check prefix for shifting pointers(either pointer of prefix or pointer of feature)
    features_quantity: Dict[str, int] = {}
    while feature_ptr < len(features) and pref_ptr < len(prefixes):
        prefix: str = prefixes[pref_ptr]
        feature: str = features[feature_ptr]
        if not feature.startswith(prefix):
            if was_previously:
                pref_ptr += 1
                was_previously = False
            else:
                feature_ptr += 1
            continue
        was_previously = True
        if features_quantity.get(prefix, None) is None:
            features_quantity[prefix] = 1
        else:
            features_quantity[prefix] += 1
        feature_ptr += 1
    return features_quantity
```

```python
all_features_quantity = get_features_quantity(features, drop_features_prefixes)
total_features: int = 0
for prefix, quantity in all_features_quantity.items():
    print(f"Для префикса {prefix} существует {quantity} признаков")
    total_features += quantity
print(f"Всего признаков с префиксами признаков, не имеющих объектов, равно {total_features}")
```

```python
drop_features_quantity = get_features_quantity(drop_features, drop_features_prefixes)
total_drop_features: int = 0
for prefix, quantity in drop_features_quantity.items():
    print(f"Для префикса {prefix} существует {quantity} признаков, не содержащих объекты")
    total_drop_features += quantity
print(f"Всего признаков, не имеющих объектов, равно {total_drop_features}")
```

```python
pref_ptr = feature_ptr = 0
filename: str = r"D:\Obfuscation\data\drop_features_classified.txt"
was_previously: bool = False

while pref_ptr < len(drop_features_prefixes) and feature_ptr < len(drop_features):
    drop_feature: str = drop_features[feature_ptr]
    prefix: str = drop_features_prefixes[pref_ptr]
        
    if not drop_feature.startswith(prefix):
        if was_previously:
            pref_ptr += 1
            was_previously = False
            ratio: str = f"Соотношение признаков с данным префиксом(не имеющие объектов/все) равно {drop_features_quantity[prefix]}/{all_features_quantity[prefix]}."
            write_file(filename, ratio)
        else:
            feature_ptr += 1
        continue
    
    if not was_previously:
        write_file(filename, f"\n{prefix}:")
    
    was_previously = True
    postfix: str = drop_feature[len(prefix):]
    write_file(filename, f"\t{postfix},")
    feature_ptr += 1

if was_previously:
    ratio: str = f"Соотношение признаков с данным префиксом(не имеющие объектов/все) равно {drop_features_quantity[prefix]}/{all_features_quantity[prefix]}."
    write_file(filename, ratio)
```
